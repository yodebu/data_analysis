```python
## Import conventions for Pandas and Numpy 
import pandas as pd
import numpy as np

## for simplicity
Panel = pd.Panel
DataFrame = pd.DataFrame
Series = pd.Series
randn = np.random.rand

## Some Pandas options
pd.set_option("html",False) ## If you don't want html outputs
pd.set_option("max_columns",30)
pd.set_option("max_rows",20)
```

```python
## init dataframes
m1 = Series(randn(100))
m2 = Series(randn(100))
m3 = Series(randn(101))
dict_of_series = {"m1": m1, "m2": m2, "m3": m3}
df1 = pd.DataFrame(data=dict_of_series)
df2 = df1 + 2
print df1
print df2
```

```python
## dict of DataFrame to Panel
data = {"item1": df1, "item2": df2}
wp1 = Panel(data)
wp1
```

```python
wp2 = Panel.from_dict(data, orient="minor")
wp2
```

```python
wp2["m1"]
```

```python
%timeit wp1["item3"] = wp1["item2"] + wp1["item1"] 
wp1["item3"]
```

```python
%timeit df3 = df1 + df2
df3
```
